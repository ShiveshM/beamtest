```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def get_charge(df):
    idxs = df['index'].unique()
    charge = []
    for idx in idxs:
        charge.append(df[df['index'] == idx]['charge'][0])
    return np.array(charge)
```

# Original Backward Configuration

## Electron data - 8 GeV

```python
store = pd.HDFStore("./bwd_1/8GeV_e.hd5")
e8_bw1 = store['df']
store.close()
```

```python
print 'Number of waveforms = {0}'.format(len(e8_bw1['index'].unique()))
_ = plt.hist(get_charge(e8_bw1),50)
```
