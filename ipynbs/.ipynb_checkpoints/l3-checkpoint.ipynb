```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
```

```python
def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))
```

```python
def multiple_gauss(x, *p):
    return gauss(x, p[0], p[1], p[2]) + gauss(x, p[3], p[4], p[5])
```

```python
INFILE = '/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/processed/l2/bwd_2/8GeV_e.hd5'
```

```python
store = pd.HDFStore(INFILE)
df = store['df']
store.close()
```

```python
ax = None
for idx in xrange(100):
    ax = df[df['index'] == idx].plot(x='isamp', y='voltage', ax=ax, color='b', legend=False)
```

```python
wv = df[df['index'] == 74]
wv.plot(x='isamp', y='voltage')
wv_fp = wv[wv['isamp'] < 150]
```

```python
# initial guess
p0 = [1., 120., 1.]
coeff, var_matrix = curve_fit(gauss, wv_fp['isamp'], wv_fp['voltage'], p0=p0, bounds=([0, 0, 0], [9999, 9999, 9999]))
print 'coeff', coeff
fit = gauss(wv['isamp'], *coeff)
wv.plot(x='isamp', y='voltage', legend=None, color='grey')
plt.plot(wv['isamp'], fit, color='red')
```

```python
# initial guess
p0 = [1., 120., 1., 1., 180., 1.]
coeff, var_matrix = curve_fit(multiple_gauss, wv['isamp'], wv['voltage'], p0=p0, bounds=([0]*6, [9000]*6))
print 'coeff', coeff
fit = multiple_gauss(wv['isamp'], *coeff)
wv.plot(x='isamp', y='voltage', legend=None, color='grey')
plt.plot(wv['isamp'], fit, color='red')
```
