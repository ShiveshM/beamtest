```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
INFILE = '/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/processed/l2/bwd_2/8GeV_e.hd5'
```

```python
store = pd.HDFStore(INFILE)
df = store['df']
store.close()
# print df
```

```python
ax = None
for idx in xrange(df['index'].max()):
    ax = df[df['index'] == idx].plot(x='isamp', y='voltage', ax=ax, color='b', legend=False)
```

```python
pre_trigger = (0, 100) # ns
threshold = 30 # mV
```

```python
df_l3 = df.copy()
df_l3['pass_l3'] = True
for idx in xrange(df['index'].max()):
    wv = df_l3.ix[df_l3['index'] == idx]
    if ((wv['isamp'] >= pre_trigger[0]) & (wv['isamp'] <= pre_trigger[1]) & (wv['voltage'] >= threshold)).any():
        df_l3.ix[df_l3['index'] == idx, 'pass_l3'] = False
```

```python
ax = None
for idx in xrange(df_l3['index'].max()):
    wv = df_l3[df_l3['index'] == idx]
    if (wv['pass_l3'] == True).any(): continue
    ax = wv.plot(x='isamp', y='voltage', ax=ax, color='b', legend=False)
```
