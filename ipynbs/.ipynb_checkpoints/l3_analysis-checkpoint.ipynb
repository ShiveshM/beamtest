```python
# Filter bad waveforms
mask = ((e_df['gauss_mu']>100) & (e_df['gauss_mu']<150) &
        (e_df['gauss_sigma']>2.5) & (e_df['gauss_sigma']<15))
print 'e wv removed = ', np.sum(~mask) / 80.
print 'e wv removed % = ', (np.sum(~mask) / 80.) / (e_df['index'].max() + 1) * 100
e_df_f = e_df[mask]
mask = ((pi_df['gauss_mu']>100) & (pi_df['gauss_mu']<150) &
        (pi_df['gauss_sigma']>2.5) & (pi_df['gauss_sigma']<15))
print 'pi wv removed = ', np.sum(~mask) / 80.
print 'pi wv removed % = ', (np.sum(~mask) / 80.) / (pi_df['index'].max() + 1) * 100
pi_df_f = pi_df[mask]
```

```python
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
def plot_bounds(x, y, alpha, ax, **kwargs):
    X = sm.add_constant(x)

    res = sm.OLS(y, X).fit()

    prstd, iv_l, iv_u = wls_prediction_std(res, alpha=alpha)
    
    ax.plot(x, iv_u, '-', **kwargs)
    ax.plot(x, iv_l, '-', **kwargs)
```

```python
e_data = []
pi_data = []
for idx in xrange(e_df_f['index'].max() + 1):
    mask_idx = e_df_f['index'] == idx
    try:
        wv = e_df_f[mask_idx]
        e_data.append([wv['charge'][0], wv['gauss_sigma'][0]])
    except: continue

for idx in xrange(pi_df_f['index'].max() + 1):
    mask_idx = pi_df_f['index'] == idx
    try:
        wv = pi_df_f[mask_idx]
        pi_data.append([wv['charge'][0], wv['gauss_sigma'][0]])
    except:
        continue
e_data = np.array(e_data)
pi_data = np.array(pi_data)

print e_data.shape
print pi_data.shape
```

```python
fontsize = 16
alpha = 1-0.68

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)

ax.set_xlim(0, 70)
ax.set_ylim(2.5, 15)
for ymaj in ax.yaxis.get_majorticklocs():
    ax.axhline(y=ymaj, ls=':', color='gray', alpha=1, linewidth=0.3)
for xmaj in ax.xaxis.get_majorticklocs():
    ax.axvline(x=xmaj, ls=':', color='gray', alpha=1, linewidth=0.3)

ax.set_xlabel(r'Charge (nVs)', fontsize=fontsize)
ax.set_ylabel(r'Primary pulse $\sigma_p$ (ns)', fontsize=fontsize)
ax.tick_params(axis='x', labelsize=fontsize)
ax.tick_params(axis='y', labelsize=fontsize)

x, y = deepcopy(e_data.T)
x_pe = x / (GAIN * EQ * RESISTANCE * 1e9)
ax.scatter(x, y, marker='o', s=10, edgecolors='red', facecolors='none',
          linewidth=0.7, label=r'$2\:{\rm GeV}\:e$', zorder=2)
plot_bounds(x, y, alpha, ax, color='r', linewidth=2)

x, y = deepcopy(pi_data.T)
x_pe = x / (GAIN * EQ * RESISTANCE * 1e9)
ax.scatter(x, y, marker='^', s=10, edgecolors='b', facecolors='none',
          linewidth=0.7, label=r'$8\:{\rm GeV}\:\pi$')
plot_bounds(x, y, alpha, ax, color='b', linewidth=2)

ax.legend(loc='upper left', prop={'size': fontsize})

fig.savefig('./images/primary_psd_charge.pdf', bbox_inches='tight', dpi=150)
```
