```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.ndimage import gaussian_filter
from collections import OrderedDict
import itertools
FONTSIZE = 16
mpl.rcParams['font.size'] = FONTSIZE
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Computer Modern, Latin Modern Roman, Bitstream Vera Serif'
mpl.rcParams['text.usetex'] = True
```

```python
def flip(items, ncol):
    return itertools.chain(*[items[i::ncol] for i in range(ncol)])
```

```python
PI_ENERGIES = [8]
E_ENERGIES = [2]
MU_ENERGIES = []
E_SIM_EXTRA = [1]

E_DATACOLOURS = {2:'red'}
PI_DATACOLOURS = {4:'blue', 8:'blue'}
MU_DATACOLOURS = {25:'black'}
E_SIMCOLOURS = {1:'green', 2:'red'}
PI_SIMCOLOURS = {4:'blue', 8:'blue'}
MU_SIMCOLOURS = {25:'black'}
E_LS = {1:'-.', 2:'--'}
PI_LS = {4:':', 8:':'}
MU_LS = {25:'-'}
```

```python
E_SIMDIR = r'/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/data/geant/{0}GeV_e.txt'
PI_SIMDIR = r'/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/data/geant/{0}GeV_pi.txt'
MU_SIMDIR = r'/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/data/geant/{0}GeV_mu.txt'

E_SIMPATHS = OrderedDict(zip(E_ENERGIES+E_SIM_EXTRA ,[E_SIMDIR.format(x) for x in E_ENERGIES+E_SIM_EXTRA]))
PI_SIMPATHS = OrderedDict(zip(PI_ENERGIES ,[PI_SIMDIR.format(x) for x in PI_ENERGIES]))
MU_SIMPATHS = OrderedDict(zip(MU_ENERGIES ,[MU_SIMDIR.format(x) for x in MU_ENERGIES]))

DOM_EFF = 0.10
PI_X_SCALE = (1/3.) #* 0.7 * 0.8
E_X_SCALE = PI_X_SCALE
MU_X_SCALE = PI_X_SCALE

E_DATADIR = r'/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/data/l3/bwd_2/{0}GeV_e.hd5'
PI_DATADIR = r'/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/data/l3/bwd_2/{0}GeV_pi.hd5'
MU_DATADIR = r'/home/shivesh/Documents/physics/PhD/beamtest/DDC2/pingu/latest/data/l3/bwd_2/{0}GeV_mu.hd5'

E_DATAPATHS = OrderedDict(zip(E_ENERGIES ,[E_DATADIR.format(x) for x in E_ENERGIES]))
PI_DATAPATHS = OrderedDict(zip(PI_ENERGIES ,[PI_DATADIR.format(x) for x in PI_ENERGIES]))
MU_DATAPATHS = OrderedDict(zip(MU_ENERGIES ,[MU_DATADIR.format(x) for x in MU_ENERGIES]))

BINNING = np.linspace(0, 80, 50)
BIN_CENTERS = (BINNING[:-1] + BINNING[1:])/2.

GAIN = 2e7
EQ = 1.6e-19 # C
# RESISTANCE = 50 # Ohms
RESISTANCE = 150 # Ohms
```

```python
E_RAWSIM = OrderedDict()
for en in E_SIMPATHS.iterkeys():
    E_RAWSIM[en] = np.genfromtxt(E_SIMPATHS[en], usecols=(0), dtype=np.float)

PI_RAWSIM = OrderedDict()
for en in PI_SIMPATHS.iterkeys():
    PI_RAWSIM[en] = np.genfromtxt(PI_SIMPATHS[en], usecols=(0), dtype=np.float)

MU_RAWSIM = OrderedDict()
for en in MU_SIMPATHS.iterkeys():
    MU_RAWSIM[en] = np.genfromtxt(MU_SIMPATHS[en], usecols=(0), dtype=np.float)
```

```python
binning = np.linspace(0, 10000, 50)
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111)

ax.set_xlim(np.min(binning), np.max(binning))
ax.set_ylim(1, 1e4)

ax.set_yscale('log')

ax.tick_params(axis='x', labelsize=FONTSIZE)
ax.tick_params(axis='y', labelsize=FONTSIZE)

ax.set_xlabel('PE', fontsize=FONTSIZE)
ax.set_ylabel('Number of entries (a.u.)', fontsize=FONTSIZE)

for en in E_RAWSIM.iterkeys():
    hist, _ = np.histogram(E_RAWSIM[en], binning)
    hist_0 = np.concatenate([[hist[0]], hist])
    
    ax.step(binning, hist_0, drawstyle='steps-pre', linestyle='--',
            color=E_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:e^-$')

for en in PI_RAWSIM.iterkeys():
    hist, _ = np.histogram(PI_RAWSIM[en], binning)
    hist_0 = np.concatenate([[hist[0]], hist])
    
    ax.step(binning, hist_0, drawstyle='steps-pre', linestyle='--',
            color=PI_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:\pi^-$')

for en in MU_RAWSIM.iterkeys():
    hist, _ = np.histogram(MU_RAWSIM[en], binning)
    hist_0 = np.concatenate([[hist[0]], hist])
    
    ax.step(binning, hist_0, drawstyle='steps-pre', linestyle='--',
            color=MU_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:\mu^-$')
    
for ymaj in ax.yaxis.get_majorticklocs():
    ax.axhline(y=ymaj, ls=':', color='gray', alpha=1, linewidth=0.3)
for xmaj in ax.xaxis.get_majorticklocs():
    ax.axvline(x=xmaj, ls=':', color='gray', alpha=1, linewidth=0.3)

ax.legend(prop=dict(size=FONTSIZE))
```

```python
E_SIM = OrderedDict()
for en in E_RAWSIM.iterkeys():
    E_SIM[en] = E_RAWSIM[en] * E_X_SCALE * DOM_EFF
    # Change from PE to nVs
    E_SIM[en] *= GAIN * EQ * RESISTANCE * 1e9

PI_SIM = OrderedDict()
for en in PI_RAWSIM.iterkeys():
    PI_SIM[en] = PI_RAWSIM[en] * PI_X_SCALE * DOM_EFF
    # Change from PE to nVs
    PI_SIM[en] *= GAIN * EQ * RESISTANCE * 1e9

MU_SIM = OrderedDict()
for en in MU_RAWSIM.iterkeys():
    MU_SIM[en] = MU_RAWSIM[en] * MU_X_SCALE * DOM_EFF
    # Change from PE to nVs
    MU_SIM[en] *= GAIN * EQ * RESISTANCE * 1e9
```

```python
binning = np.linspace(0, 200, 50)
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111)

ax.set_xlim(np.min(binning), np.max(binning))
ax.set_ylim(1, 1e4)

ax.set_yscale('log')

ax.tick_params(axis='x', labelsize=FONTSIZE)
ax.tick_params(axis='y', labelsize=FONTSIZE)

ax.set_xlabel('Charge (nVs)', fontsize=FONTSIZE)
ax.set_ylabel('Number of entries (a.u.)', fontsize=FONTSIZE)

for en in E_SIM.iterkeys():
    hist, _ = np.histogram(E_SIM[en], binning)
    hist_0 = np.concatenate([[hist[0]], hist])
    
    ax.step(binning, hist_0, drawstyle='steps-pre', linestyle='--',
            color=E_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:e^-$')

for en in PI_SIM.iterkeys():
    hist, _ = np.histogram(PI_SIM[en], binning)
    hist_0 = np.concatenate([[hist[0]], hist])
    
    ax.step(binning, hist_0, drawstyle='steps-pre', linestyle='--',
            color=PI_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:\pi^-$')

for en in MU_SIM.iterkeys():
    hist, _ = np.histogram(MU_SIM[en], binning)
    hist_0 = np.concatenate([[hist[0]], hist])
    
    ax.step(binning, hist_0, drawstyle='steps-pre', linestyle='--',
            color=MU_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:\mu^-$')
    
for ymaj in ax.yaxis.get_majorticklocs():
    ax.axhline(y=ymaj, ls=':', color='gray', alpha=1, linewidth=0.3)
for xmaj in ax.xaxis.get_majorticklocs():
    ax.axvline(x=xmaj, ls=':', color='gray', alpha=1, linewidth=0.3)

ax.legend(prop=dict(size=FONTSIZE))
```

```python
E_DF = OrderedDict()
for en in E_DATAPATHS.iterkeys():
    store = pd.HDFStore(E_DATAPATHS[en])
    E_DF[en] = store['df']
    store.close()

PI_DF = OrderedDict()
for en in PI_DATAPATHS.iterkeys():
    store = pd.HDFStore(PI_DATAPATHS[en])
    PI_DF[en] = store['df']
    store.close()
    
MU_DF = OrderedDict()
for en in MU_DATAPATHS.iterkeys():
    store = pd.HDFStore(MU_DATAPATHS[en])
    MU_DF[en] = store['df']
    store.close()
```

```python
E_DATA = OrderedDict()
for en in E_DF.iterkeys():
    E_DATA[en] = []
    for idx in xrange(E_DF[en]['index'].max() + 1):
        mask_idx = E_DF[en]['index'] == idx
        wv = E_DF[en][mask_idx]
        E_DATA[en].append(wv['charge'][0])
        #E_DATA[en].append(wv['gauss_int'][0])
    E_DATA[en] = np.array(E_DATA[en])

PI_DATA = OrderedDict()
for en in PI_DF.iterkeys():
    PI_DATA[en] = []
    for idx in xrange(PI_DF[en]['index'].max() + 1):
        mask_idx = PI_DF[en]['index'] == idx
        wv = PI_DF[en][mask_idx]
        PI_DATA[en].append(wv['charge'][0])
        #PI_DATA[en].append(wv['gauss_int'][0])
    PI_DATA[en] = np.array(PI_DATA[en])

MU_DATA = OrderedDict()
for en in MU_DF.iterkeys():
    MU_DATA[en] = []
    for idx in xrange(MU_DF[en]['index'].max() + 1):
        mask_idx = MU_DF[en]['index'] == idx
        wv = MU_DF[en][mask_idx]
        MU_DATA[en].append(wv['charge'][0])
        #MU_DATA[en].append(wv['gauss_int'][0])
    MU_DATA[en] = np.array(MU_DATA[en])
```

```python
e_simweight = OrderedDict()
for en in E_ENERGIES:
    e_simweight[en] = [len(E_DATA[en]) / float(len(E_SIM[en]))]*len(E_SIM[en])
for en in E_SIM_EXTRA:
    e_simweight[en] = [1.]*len(E_SIM[en])

pi_simweight = OrderedDict()
for en in PI_ENERGIES:
    pi_simweight[en] = [len(PI_DATA[en]) / float(len(PI_SIM[en]))]*len(PI_SIM[en])

mu_simweight = OrderedDict()
for en in MU_ENERGIES:
    mu_simweight[en] = [len(MU_DATA[en]) / float(len(MU_SIM[en]))]*len(MU_SIM[en])
```

```python
for en in E_SIM.iterkeys():
    is_data = en in E_DATA.keys()
    fig = plt.figure(figsize=(7, 6))
    ax = fig.add_subplot(111)

    ax.set_xlim(np.min(BINNING), np.max(BINNING))
    ax.set_ylim(0, 100)

    for ymaj in ax.yaxis.get_majorticklocs():
        ax.axhline(y=ymaj, ls=':', color='gray', alpha=1, linewidth=0.3)
    for xmaj in ax.xaxis.get_majorticklocs():
        ax.axvline(x=xmaj, ls=':', color='gray', alpha=1, linewidth=0.3)

    ax.tick_params(axis='x', labelsize=FONTSIZE)
    ax.tick_params(axis='y', labelsize=FONTSIZE)

    ax.set_xlabel('Charge (nVs)', fontsize=FONTSIZE)
    ax.set_ylabel('Number of entries (a.u.)', fontsize=FONTSIZE)

    hist_sim, _ = np.histogram(E_SIM[en], BINNING, weights=np.array(e_simweight[en]))
    if is_data: hist_data, _ = np.histogram(E_DATA[en], BINNING)

    hist_sim_0 = np.concatenate([[hist_sim[0]], hist_sim])
    if is_data: hist_data_0 = np.concatenate([[hist_data[0]], hist_data])

    #if is_data: ax.step(BINNING, hist_data_0, drawstyle='steps-pre', linestyle='-',
    #                    color=E_DATACOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:e^-\:{\rm data}$')
    if is_data:
        ax.scatter(BIN_CENTERS, hist_data, marker='o', s=10, color=E_DATACOLOURS[en],
          linewidth=0.7, label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:e^-\:{\rm data}$')
        ax.errorbar(
            BIN_CENTERS, hist_data, color=E_DATACOLOURS[en],
            xerr=0, yerr=np.sqrt(hist_data), capsize=1.3,
            alpha=1, linestyle='None', markersize=2, linewidth=1.2
        )
    ax.step(BINNING, hist_sim_0, drawstyle='steps-pre', linestyle='--',
            color=E_SIMCOLOURS[en], label=r'$'+'{0}'.format(en)+r'~{\rm GeV}\:e^-\:{\rm sim}$')

    ax.legend(prop=dict(size=FONTSIZE))
```
